```python
text = open("task.txt", "r", encoding='utf-8').read()
```

```python
import base64
```

```python
# расшифровали текст
first_iter = base64.b64decode(text)
second_iter = base64.b64decode(first_iter)
print(second_iter.decode('utf-8'))
```
